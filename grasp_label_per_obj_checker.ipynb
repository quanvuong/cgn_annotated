```python
import numpy as np
from pathlib import Path
```

##### per-obj grasps and labels from original file (generated in mobile_pnp)

```python
from contact_graspnet.data import PointCloudReader
from utils import load_label_from_bfs, get_path_to_recon_mesh

mobile_pnp_dir = Path('/home/robin/Projects/mobile_pnp_branches/mobile_pnp_collapsing_grasp_success_heuristic_fix_for_label_generation/')

num_test_scene = 30 

scene_level_grasps = dict()

for env_idx in range(num_test_scene):

    recon_env_name = f'ArmGraspObjectClutterReconstruction_{env_idx}-v0'

    pcreader = PointCloudReader(root_folder=None)

    obj_name_to_bfs_label = load_label_from_bfs(
        recon_env_name, mobile_pnp_dir
    )

    scene_level_grasps[env_idx] = obj_name_to_bfs_label
```

```python
def print_grasp_info(scene_level_grasps):
    scene_total_grasps_info = dict()
    for scene_idx, scene_grasp_info in scene_level_grasps.items():
        print(f'Scene_{scene_idx}')
        total_scene_grasps = 0
        total_scene_success_grasps = 0
        for obj_name in sorted(scene_grasp_info.keys()):
            obj_grasp_info = scene_grasp_info[obj_name]
            if obj_grasp_info == None:
                print(f'\t {obj_name}')
                print(f'\t \t total grasps: 0')
                print(f'\t \t success grasps: 0')
            else:
                print(f'\t {obj_name}')
                print(f'\t \t total grasps: {len(obj_grasp_info["grasp_success_labels"])}')
                print(f'\t \t success grasps: {np.sum(obj_grasp_info["grasp_success_labels"])}')
                total_scene_grasps += len(obj_grasp_info["grasp_success_labels"])
                total_scene_success_grasps += np.sum(obj_grasp_info["grasp_success_labels"])
        scene_total_grasps_info[scene_idx] = {'total': total_scene_grasps, 'success':total_scene_success_grasps}
    return scene_total_grasps_info
```

```python
scene_total_grasp_info =  print_grasp_info(scene_level_grasps)
```
