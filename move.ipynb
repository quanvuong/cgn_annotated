```python
import os
import glob

obj_file = os.listdir()

obj_file = [
    f for f in obj_file if f.endswith('.obj')
]
```

```python
from pathlib import Path
import shutil

for f in obj_file:

    new_f_name = f.replace(
        'temp.', ''
    ).replace(
        '.watertight', ''
    )

    print(new_f_name)

    move_to = Path('/home/tele_mani_u20/code_proj/cgn/June_4/contact_graspnet/acronym/meshes')

    move_to.mkdir(exist_ok=True, parents=True)

    shutil.move(f, move_to / new_f_name)
```
